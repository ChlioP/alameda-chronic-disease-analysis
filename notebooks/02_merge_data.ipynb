```python
import pandas as pd
```

```python
income18 = pd.read_csv('ACSDP5Y2018.DP03-Data.csv')
income22 = pd.read_csv('ACSDP5Y2022.DP03-Data.csv')
housing18 = pd.read_csv('ACSDP5Y2018.DP04-Data.csv')
housing22 = pd.read_csv('ACSDP5Y2022.DP04-Data.csv')
merge18 = pd.read_csv('merge18.csv')
merge22 = pd.read_csv('merge22.csv')
```

```python
income18 = income18[['NAME', 'DP03_0062E', 'DP03_0119PE']].copy()
income18 = income18[income18['NAME'] == 'Alameda County, California']
income18 = income18.rename(columns={'DP03_0062E': 'median_income_2018', 'DP03_0119PE': 'poverty_pct_2018'})
income18['County'] = 'Alameda'
income18
```

```python
income22 = income22[['NAME', 'DP03_0062E', 'DP03_0119PE']].copy()
income22 = income22[income22['NAME'] == 'Alameda County, California']
income22 = income22.rename(columns={'DP03_0062E': 'median_income_2022', 'DP03_0119PE': 'poverty_pct_2022'})
income22['County'] = 'Alameda'
income22
```

```python
housing18 = housing18[['NAME', 'DP04_0136E', 'DP04_0101E']].copy()
housing18 = housing18[housing18['NAME'] == 'Alameda County, California']
housing18 = housing18.rename(columns={'DP04_0136E': 'median_rent_2018', 'DP04_0101E': 'median_owner_cost_2018' })
housing18['County'] = 'Alameda'
housing18
```

```python
housing22 = housing22[['NAME', 'DP04_0136E', 'DP04_0101E']].copy()
housing22 = housing22[housing22['NAME'] == 'Alameda County, California']
housing22 = housing22.rename(columns={'DP04_0136E': 'median_rent_2022', 'DP04_0101E': 'median_owner_cost_2022' })
housing22['County'] = 'Alameda'
housing22
```

```python
# Merge into the final dataset
merge18 = merge18.merge(housing18[['County', 'median_rent_2018', 'median_owner_cost_2018']], on='County', how='left')
merge22 = merge22.merge(housing22[['County', 'median_rent_2022', 'median_owner_cost_2022']], on='County', how='left')
```

```python
merge18 = merge18.merge(income18[['County', 'median_income_2018', 'poverty_pct_2018']], on='County', how='left')
merge22 = merge22.merge(income22[['County', 'median_income_2022', 'poverty_pct_2022']], on='County', how='left')
```

```python
merge18.to_csv('merge18.csv', index=False)
merge22.to_csv('merge22.csv', index=False)
```
